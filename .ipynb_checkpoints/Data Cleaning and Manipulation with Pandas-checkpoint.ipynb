```python
import io
import re
import pandas as pd
import numpy as np
data = pd.read_csv("C:/attacks.csv", encoding = "ISO-8859-1")
```

```python
#ELIMINAR COLUMNAS FINALES SIN DATOS
data = data.drop(['Unnamed: 22','Unnamed: 23'], axis = 1)

#ELIMINAR TODAS LAS FILAS QUE TENGAN NA, TODAS LAS QUE COMIENCEN CON 0 O CON **
data = data.dropna(how='all')
data = data.drop(data[data['Case Number']=='0'].index)
data = data.drop(data[data['Case Number']=='xx'].index)

data = data.rename(columns={'Case Number' : 'Case_Number'})

for i in data:
    data[i] = data[i].fillna('Unkown')
    
for i in data:
    data[i] = data [i].astype(str).str.replace(r'\W', '')

#Renombrar CASE_NUMBER
# Registro vacio en CASE_NUMBER colocar ocmo Unknown
# VALORES VACIOS EN TYPE COLOCAR COMO "Unknown"
#CAMBIAR POR BOATING: BOAT Y BOATONG
#CAMBIAR POR RED SEA / INDIAN OCEAN: RED SEA Y RED SEA?
#TODO MAYUSC EN COUNTRY
#ESTANDARIZAR COL SEX
    
data.Type = data.Type.replace({"boat" :"boating" , 'Boatomg' : 'boating'})
data.Country = data.Country.replace({"RED SEA?" :"RED SEA / INDIAN OCEAN" , "RED SEA? / INDIAN OCEAN" : "RED SEA / INDIAN OCEAN"})
data.Country = data['Country'].str.upper()
data = data.rename(columns={'Sex ' : 'Sex'})
data.Sex = data.Sex.replace({"." :"N/A" , "lli" :"M", "N" : "M"})
data['Sex']=data['Sex'].fillna('Unkown')
data["Name"] = data["Name"].str.replace('""','')
#data['Name']= data['Name'].str.replace(r"\W","")
data.sort_values('Case_Number', ascending=1)
data.head()
```

```python
data.to_csv('sharkattack_clean.csv', index=0, encoding="utf-8")
```
